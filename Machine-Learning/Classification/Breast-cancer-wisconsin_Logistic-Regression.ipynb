# Import the dataset

```python
dataset_df = pd.read_csv('breast-cancer-wisconsin-data.csv', na_values = '?')   # The symbol '?' is assumed as missing values or NaN.

dataset_df
```

# Data pre-processing

```python
dataset_df.info()
```

```python
# Checking the missing values.

dataset_df.isnull().sum()

# The column 'Bare Nuclei' has 16 missing values or nan values which are '?' symbol.
```

```python
# Drop the rows which has at least one missing value.

dataset_df.dropna(inplace = True)

dataset_df
```

```python
dataset_df.isnull().sum()

# Now no missing values in any column.
```

```python
dataset_df.describe()
```

```python
X = dataset_df.iloc[:, :-1].values   # Matrix of features.
y = dataset_df.iloc[:, -1].values   # Dependent variable or target variable vector.
```

```python
X
```

```python
X.shape
```

```python
y
```

```python
y.shape
```

# Splitting into train and test sets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

# Training the logistic regression model on the training set

```python
from sklearn.linear_model import LogisticRegression

classifier_log_reg = LogisticRegression(random_state = 0)

classifier_log_reg.fit(X_train, y_train)
```

# Prediction on the training set

```python
y_pred_train = classifier_log_reg.predict(X_train)
y_pred_train
```

```python
for a, b in zip(y_pred_train, y_train):
    print(a, b)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

print('The confusion matrix on the training set is: \n', confusion_matrix(y_train, y_pred_train))
print('The accuracy of the logistic regression model on the training set is =', accuracy_score(y_train, y_pred_train))
```

# Prediction on the test set

```python
y_pred_test = classifier_log_reg.predict(X_test)
y_pred_test
```

```python
for a, b in zip(y_pred_test, y_test):
    print(a, b)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

print('The confusion matrix on the test set is: \n', confusion_matrix(y_test, y_pred_test))
print('The accuracy of the logistic regression model on the test set is =', accuracy_score(y_test, y_pred_test))
```
